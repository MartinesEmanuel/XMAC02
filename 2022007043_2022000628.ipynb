O teste de hipótese que realizei foi baseado no dataset "iris". ***Minha hipótese nula foi de que não há diferença significativa entre os comprimentos das pétalas das espécies "setosa" e "versicolor" de íris.***


Para realizar o teste, selecionei uma amostra aleatória do dataset que não excedesse 20% do tamanho total. Utilizei essa amostra para comparar as médias dos comprimentos das pétalas das duas espécies e verifiquei se havia diferença significativa entre elas.


Usei o *teste t* de Student para essa comparação. Ele fornece uma *estatística t *e um *valor-p.* O *valor-p *é a probabilidade de obter uma diferença tão grande ou maior entre as médias dos comprimentos das pétalas, assumindo que a hipótese nula seja verdadeira. \


Com um nível de significância de 0,05 (ou 5%), realizei o teste de hipótese. Se o valor-p fosse menor que 0,05, poderíamos rejeitar a hipótese nula e concluir que existia uma diferença significativa nos comprimentos das pétalas entre as espécies "setosa" e "versicolor". Caso contrário, não teríamos evidências suficientes para rejeitar a hipótese nula.


```python
import seaborn as sns
import pandas as pd
from scipy import stats


iris = sns.load_dataset("iris")
iris.head()

```

```python
iris.describe()
```

```python
amostra = iris[iris["species"].isin(["setosa", "versicolor"])].sample(frac=0.2, random_state=42)
amostra.head()
```

```python
setosa = amostra[amostra["species"] == "setosa"]["petal_length"]
versicolor = amostra[amostra["species"] == "versicolor"]["petal_length"]

t_statistic, p_value = stats.ttest_ind(setosa, versicolor)
t_statistic, p_value
```

 ***Podemos rejeitar a hipótese nula.*** Isso significa que há evidências estatísticas para concluir que os comprimentos das pétalas das espécies "setosa" e "versicolor" de íris são significativamente diferentes.
